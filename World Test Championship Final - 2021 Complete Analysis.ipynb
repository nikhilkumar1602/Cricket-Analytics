```python
indteamruns = df[df.batting_team == 'India']
```

```python
indteamruns.tail()
```

```python
indbatsmen = pd.DataFrame(indteamruns.groupby('striker')['runs_off_bat'].sum()).reset_index()
```

```python
indbatsmen1 = pd.DataFrame(indteamruns.groupby('striker')['ball'].count()).reset_index()
```

```python
indbatsmen1.head(12)
```

```python
indbatsmen.head(12)
```

```python
indbatting = indbatsmen.merge(indbatsmen1, on = 'striker', how = 'left')
```

```python
indbatting.head(12)
```

```python
for player in indbatting['striker']:
    strike_rate = 100*(indbatting['runs_off_bat']/indbatting['ball'])
    
    
```

```python
strike_rate.head(12)
```

```python
indbatting['strike_rate'] = strike_rate
```

```python
for player in indbatting['striker']:
    plycontrib = 100*(indbatting['runs_off_bat']/sum(indbatting['runs_off_bat']))
```

```python
indbatting['players_contibution'] = plycontrib
```
